```python
# Safety check: if a password variable pwd is at least 8 characters long and contains at least one digit, print "Password is strong"; otherwise print "Password is weak".
pwd = input('Enter the password')

if(len(pwd) >= 8 ) & any(ch.isdigit() for ch in pwd):
    print('Password is strong')
else:
    print('Password is weak')
```

```python
# Safety check: if a password variable pwd is at least 8 characters long and contains at least one digit,one special char,one upper case, print "Password is strong"; otherwise print "Password is weak".
pwd = input("Enter the password: ")

has_length   = len(pwd) >= 8
has_digit    = any(ch.isdigit() for ch in pwd)
has_upper    = any(ch.isupper() for ch in pwd)
has_special  = any(ch in "!@#$%^&*()-_=+[]{};:'\",.<>/?|`~" for ch in pwd)

if has_length and has_digit and has_upper and has_special:
    print("Password is strong")
else:
    print("Password is weak")
```

```python
# Print the largest and smallest digit in a given number using while loop.
number = 489273615
largest = 0
smallest = 9
n = number

while n > 0:
      digit = n % 10
      if digit > largest:
          largest = digit
      if digit < smallest:
          smallest = digit
      n = n//10

print('Largest digit:',largest)
print('smallest digit:',smallest)
```

```python
# Create and print a number series (Fibonacci sequence) using loops.
n = int(input('enter the number :'))
first = 0
second = 1
count = 0
print('Fibonacci sequence')
while n > count:
      print(first, end = ' ')
      fibo_seq = first + second
      first = second
      second  = fibo_seq
      count = count + 1               
```

```python
# Count the number of times a specific element appears in a list using a loop.
my_list = [10,20,30,20,40,20,50]
element_to_count = 20
count = 0
for item in my_list:
    if item == element_to_count:
        count = count + 1

print(f'{element_to_count} appear {count} time in the list.')
```

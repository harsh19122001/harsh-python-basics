```python
def count(lst):
    even = 0
    odd = 0
    for i in lst:
        if i%2 == 0:
            even += 1
        else:
            odd += 1
    return even,odd
lst = [1,2,3,4,8,9,10]
lst

even,odd = count(lst)

print(even)
print(odd)
```

```python
def fib(n):
    a = 0
    b = 1

    print(a)
    print(b)

    for i in range(0,n):
        c = a + b
        a = b
        b = c

        print(c)

fib(10)
```

```python
def fact(n):
    f = 1
    for i in range(1 , n+1):
        f = f * i

    return f

x = 5
result = fact(x)
print(result)
```

```python
# Lets take one list & i want to find the list of even numbers
def is_even(n):
    return n % 2 == 0

nums = [3,2,6,8,4,6,2,9]

evens = list(filter(is_even , nums))
print(evens)
```

```python
# Q8 Function to count digits in a number.
def count_digit(num):
    n = abs(num) # it handles negative numbers
    if num == 0:
       return 1  # 0 has 1 digit
    count = 0
    while num>0:
        num = num//10 # remove last digit
        count += 1
    return count
count_digit(100)
```

```python
# Q9 Function to reverse a number.
def reverse_num(num):
    reverse = 0
    while num > 0:
        lastdigit = num % 10
        reverse = lastdigit + reverse*10
        num = num//10
    return reverse
reverse_num(456)
```

```python
# Q10 Function to check whether a number is palindrome.
num = int(input('enter the num : '))
def palindrome(num):
    original = num
    reverse = 0
    while num>0:
        lastdigit = num%10
        reverse = reverse * 10 + lastdigit
        num = num//10
        
    if original == reverse:
         print('Number is palindrome')
    else:
         print('Number is not palindrome')
palindrome(num)
```

```python
# Q11 Function that returns the largest of three numbers.
def largest(A,B,C):
    if (A > B) & (A > C):
       print('A is largest')
    elif (B > A) & (B > C):
        print('B is largest')
    else:
        print('C is largest')
largest(100,200,50)
```

```python
# Q12 Function to find the sum of digits of a number.
def sum_digit(num):
    SumOfDigit = 0
    while num>0:
        lastdigit = num % 10
        SumOfDigit = SumOfDigit + lastdigit
        num = num//10
    return SumOfDigit
sum_digit(4444)
```

```python
# Q14 Function to check whether a number is prime.
import math
def isprime(n):
    if n == 2:
        return False

    for i in range(2,n):
        if n%i == 0:
            return False

    return True
isprime(8)
```
